```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
pd.set_option("display.max_rows", 15)
```

#### .READ_CSV() (read the dataset)

```python
df = pd.read_csv(r"..\..\..\..\..\..\Downloads\archive\fitness_analysis.csv")
df.head()
```

#### ------- CLEANING THE DATA -------

#### DF.RENAME() (rename columns name) 
#### DF.HEAD() ( show 5 first rows )

```python
df.rename(columns={"Timestamp":"Total_Date", "Your name ":"Name", "Your gender ":"Gender", "Your age ":"Age_Category", "How important is exercise to you ?":"Exercise_Importance",
                   "How do you describe your current level of fitness ?":"Fitness_Level", "How often do you exercise?":"Times_Week", 
                   "What barriers, if any, prevent you from exercising more regularly?           (Please select all that apply)":"Barriers_Practice",
                   "What form(s) of exercise do you currently participate in ?                        (Please select all that apply)":"Type_Exercise", "Do you exercise ___________ ?":"How_Exercise",
                   "What time if the day do you prefer to exercise?":"When_Exercise", "How long do you spend exercising per day ?":"How_Long_By_Day (Min)",
                   "Would you say you eat a healthy balanced diet ?":"Healthy_Diet", "What prevents you from eating a healthy balanced diet, If any?                         (Please select all that apply)":"Diet_Mistakes",
                   "How healthy do you consider yourself?":"Healthy_Score", "Have you ever recommended your friends to follow a fitness routine?":"Push_Others_To_Exercise",
                   "Have you ever purchased a fitness equipment?":"Equipment_Purchased", "What motivates you to exercise?         (Please select all that applies )":"Exercise_Motivation"},
                   inplace=True)
df.head()
```

#### PD.TO_DATETIME() ( convert column to pandas datetime object )
#### DF.DROP() ( drop column from dataset )

```python
df["Date"] = pd.to_datetime(df["Total_Date"]).dt.date
df["Time"] = pd.to_datetime(df["Total_Date"]).dt.strftime("%H:%M")
df.drop("Total_Date", axis=1, inplace=True)
df.head()
```

#### .ISNA().SUM() ( know the number of NULL values )
#### .UNIQUE() ( know the unique values in a column )
#### .VALUE_COUNTS() ( return the count of unique values in a column )

```python
df["Gender"].isna().sum()
df["Gender"].unique()

df["Age_Category"].value_counts()
df["Age_Category"] = df["Age_Category"].str.replace(" to ", "/", regex=False).str.replace(" and above", "+", regex=False).str.strip()
df.head()
```

#### NP.WHERE ( replace values )

```python
df["Times_Week"].value_counts()
df["Times_Week"] = np.where(df["Times_Week"] == "Never", 0, 
                           np.where(df["Times_Week"] == "1 to 2 times a week", 1.5,
                           np.where(df["Times_Week"] == "2 to 3 times a week", 2.5,
                           np.where(df["Times_Week"] == "3 to 4 times a week", 3.5,
                           np.where(df["Times_Week"] == "5 to 6 times a week", 5.5,
                           np.where(df["Times_Week"] == "Everyday", 7, df["Times_Week"]))))))
df.head()
```

```python
df["Barriers_Practice"] = [each.split(";")[0] for each in df["Barriers_Practice"]]

df.drop("Type_Exercise", axis=1, inplace=True)

df["When_Exercise"] = df["When_Exercise"].str.replace("Early morning", "Morning", regex=False)
df.head()
```

```python
df["Diet_Mistakes"].unique()
df["Diet_Mistakes"] = [each.split(";")[0] for each in df["Diet_Mistakes"]]
df.head()
```

```python
df.drop("Exercise_Motivation", axis=1, inplace=True)
df.head()
```

#### DF.DUPLICATED().SUM() ( how many duplicates values in a column )
#### DF.DROP_DUPLICATES() ( drop duplicates values in a column )

```python
df["Name"].duplicated(keep="first").sum()
```

```python
df.drop_duplicates(subset=["Name"], keep="first", inplace=True)
```

#### ------- ANALYZING THE DATA -------

```python
# Analyze the average exercise frequency between men and womens
```

```python
mean_times = df.groupby(["Gender"], as_index=False)["Times_Week"].mean()
mean_times = mean_times["Times_Week"].apply(lambda x:round(x, 1))
mean_times = mean_times.to_frame()
mean_times.rename(index={0:"Female", 1:"Male"}, inplace=True)
mean_times
```

```python
# Analyze the time spent exercising between men and womens
```

```python
male_how_long = df.loc[df["Gender"] == "Male"]["How_Long_By_Day (Min)"].value_counts().sort_index()
male_how_long
```

```python
female_how_long = df.loc[df["Gender"] == "Female"]["How_Long_By_Day (Min)"].value_counts().sort_index()
female_how_long
```

#### ------- VISUALIZATION -------

```python
plot_objects = plt.subplots(figsize=(10, 6))
fig, ax = plot_objects

ax.set_facecolor("grey")
ax.patch.set_alpha(0.3)

ax1 = ax.bar(np.arange(len(male_how_long.index)) - 0.3 / 2,
             male_how_long.values, color="#4556CF",
             edgecolor="black", linewidth=2, width=0.3)

ax2 = ax.bar(np.arange(len(female_how_long.index)) + 0.3 / 2,
             female_how_long.values, color="#8929CA",
             edgecolor="black", linewidth=2, width=0.3)

ax.legend(["Male", "Female"])

plt.xticks([0, 1, 2, 3, 4], ["1 hour", "2 hours", "3 hours and above", "30 minutes", "I don't really exercise"])

ax.bar_label(ax1, padding=2)
ax.bar_label(ax2, padding=2)

ax.set_title("Male / Female duration exercise statistics", fontsize=15, fontfamily="monospace")
ax.set_ylabel("Count", fontsize=12)
ax.set_xlabel("Duration", fontsize=12)
```

```python
grouped_diet = df.groupby(["Diet_Mistakes"])["Gender"].apply(list)
grouped_diet = grouped_diet.to_frame()
grouped_diet["Count_Male"] = grouped_diet["Gender"].apply(lambda x:x.count("Male") if isinstance(x, list) else x)
grouped_diet["Count_Female"] = grouped_diet["Gender"].apply(lambda x:x.count("Female") if isinstance(x, list) else x)
grouped_diet.drop("Gender", axis=1, inplace=True)

top5_male = grouped_diet["Count_Male"].sort_values(ascending=False)[:5].to_frame()
top5_male

top5_female = grouped_diet["Count_Female"].sort_values(ascending=False)[:5].to_frame()
top5_female
```

```python
plot_objects = plt.subplots(nrows=2, ncols=1, figsize=(7, 14))
fig, (ax1, ax2) = plot_objects

ax1.pie(top5_male["Count_Male"], shadow=True, explode=[0.1, 0, 0, 0, 0], labels=top5_male.index,
        autopct="%1.1f%%", textprops={"fontsize":10})

ax1.set_title("TOP 5 MALE DIET MISTAKES", fontsize=14, fontfamily="monospace")
ax2.set_title("TOP 5 FEMALE DIET MISTAKES", fontsize=14, fontfamily="monospace")

ax2.pie(top5_female["Count_Female"], shadow=True, explode=[0.1, 0, 0, 0, 0], labels=top5_female.index,
        autopct="%1.1f%%")
```

```python
age_when = df.groupby(["Age_Category"])["When_Exercise"].apply(list)
age_when = age_when.to_frame()
age_when = age_when.explode("When_Exercise")
age_when
```

```python
plot_objects = plt.subplots(figsize=(10, 6))
fig, ax = plot_objects

ax.set_facecolor("grey")
ax.patch.set_alpha(0.4)

ax1 = sns.countplot(y=age_when["When_Exercise"], hue=age_when.index, palette="Reds",
                    edgecolor="black", linewidth=0.5)

ax.legend(bbox_to_anchor=(0.97, 0.4))

ax.set_title("Countplot time of the day by age category", fontsize=16)
ax.set_ylabel("Time of the day", fontsize=14)
ax.set_xlabel("Count", fontsize=14)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
```

```python
plot_objects = plt.subplots(figsize=(10, 6))
fig, ax = plot_objects

ax.grid(which="major", axis="x", color="black", linestyle="-.", alpha=0.4)
ax.set_facecolor("grey")
ax.patch.set_alpha(0.3)

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["bottom"].set_visible(False)

box1 = ax.boxplot(df["Times_Week"], patch_artist=True, vert=False, 
                  whiskerprops={"linewidth":2},
                  boxprops={"linewidth":2},
                  medianprops={"color":"red", "linewidth":2},
                  capprops={"linewidth":2},
                  showfliers=False, notch=True)

ax.set_xlim(-1, 8)



for box in box1["boxes"]:
    box.set_facecolor("#BCD482")
    
ax.tick_params(axis="y", which="major", top=False, right=False, bottom=False, left=False,
               labelbottom=False, labelleft=False)

ax.set_xlabel("Count of days of week", fontsize=13)
plt.xticks([0, 1, 2, 3, 4, 5, 6, 7], fontsize=11)

ax.set_title("Boxplot times week exercising", fontsize=15)
```
